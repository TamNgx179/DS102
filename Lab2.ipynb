BÃ i 1:Sá»­ dá»¥ng Numpy xÃ¢y dá»±ng phÆ°Æ¡ng phÃ¡p Logistic Regression vÃ  Gradient Descent. Tiáº¿n hÃ nh huáº¥n luyá»‡n phÆ°Æ¡ng phÃ¡p Logistic Regression vÃ  trá»±c quan hoÃ¡ giÃ¡ trá»‹ cá»§a hÃ m loss trong toÃ n bá»™ quÃ¡ trÃ¬nh huáº¥n luyá»‡n.

```python
def convert_log(target:str) -> int:
    if target == 'Graduate':
        return 0
    else:
        return 1
```

```python
df['Target'] = df['Target'].apply(convert_log)
```

```python
data = df.to_numpy()
data_train, data_dev, data_test = np.split(data,indices_or_sections=[int(0.8*len(data)),int(0.9*len(data))])
data_train.shape,data_dev.shape, data_test.shape
```

```python
X_train = data_train[:,:-1]
y_train = data_train[:,-1].reshape(-1,1)
X_dev = data_dev[:,:-1]
y_dev = data_dev[:,-1].reshape(-1,1)
X_test = data_dev[:,:-1]
y_test = data_dev[:,-1].reshape(-1,1)
```

```python
class Logistic_Regression():
    def __init__(self, epoch, lr):
        self.epoch=epoch
        self.lr = lr
        self.w = None
        self.train_loss= []
        self.val_loss = []
        self.metrics = []
    def sigmoid(self,z):
        return 1/ (1+np.exp(-z) )   
    def loss_func(self,y,y_hat):
        return -(y*np.log(y_hat+1e-6)+(1-y)*np.log(1-y_hat+1e-6)).mean()
    def accuracy(self,y,y_hat):
        y_hat = np.where(y_hat>0.5,1,0)
        return (y_hat==y).mean()
    def predict(self,X):
        z = X @ self.w
        y_hat = self.sigmoid(z)
        return y_hat
    def fit(self,X_train,X_dev,y_train,y_dev):
        n = X_train.shape[0]
        d = X_train.shape[1]
        #weights initialization as a column vector of zeros
        self.w = np.zeros((d,1))
        for i in range(self.epoch):
            #foward pass
            y_hat = self.predict(X_train)
            #calculate accuracy
            acc = self.accuracy(y_train,y_hat)
            #train loss 
            loss = self.loss_func(y_train,y_hat)
            self.train_loss.append(loss)
            #val loss
            y_val_hat = self.predict(X_dev)
            val_loss = self.loss_func(y_dev,y_val_hat)
            self.val_loss.append(val_loss)
            #gradient
            distance = y_hat - y_train
            grad = (X_train.T @ distance)/n
            #weights update
            self.w -=  self.lr * grad
            
            self.metrics.append(acc)
            print(f'\nEPOCH {i + 1}:\tTraining loss: {loss:.3f}\tValidation loss: {val_loss:.3f}')
            
```

```python
lr = Logistic_Regression(epoch=1000,lr=0.0001)
```

```python
lr.fit(X_train,X_dev,y_train,y_dev)
```

```python
y_test_predict = lr.predict(X_test)
test_accuracy = lr.accuracy(y_test,y_test_predict)
print(test_accuracy)
```

```python
plt.figure(figsize=(10,5))
plt.plot(list(range(1, lr.epoch+1)), lr.train_loss, color = 'red' , label = 'loss')
plt.plot(list(range(1, lr.epoch+1)), lr.val_loss, color = 'blue' , label = 'val_loss')
plt.xlabel('Epoch') 
plt.ylabel('Loss')
plt.title('Loss over epochs in training')
plt.legend()
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.plot(list(range(1, lr.epoch+1)), lr.metrics, color = 'red' , label = 'accuracy')
plt.xlabel('Epoch') 
plt.ylabel('Accuracy')
plt.title('Accuracy over epochs in training')
plt.legend()
plt.grid()
plt.show()
```

BÃ i 2: Sá»­ dá»¥ng Numpy xÃ¢y dá»±ng phÆ°Æ¡ng phÃ¡p Softmax Regression vÃ  Gradient Descent. Tiáº¿n hÃ nh huáº¥n luyá»‡n phÆ°Æ¡ng phÃ¡p Logistic Regression vÃ  trá»±c quan hoÃ¡ giÃ¡ trá»‹ cá»§a hÃ m loss trong toÃ n bá»™ quÃ¡ trÃ¬nh huáº¥n luyá»‡n.

```python
df1 = pd.get_dummies(df1, columns=['Target']).astype(int)
df1
```

```python
data = df1.to_numpy()
data_train, data_dev, data_test = np.split(data,indices_or_sections=[int(0.8*len(data)),int(0.9*len(data))])
print(data_train.shape,data_dev.shape, data_test.shape)
X_train1 = data_train[:,:-3]
y_train1 = data_train[:,36:]
X_dev1 = data_dev[:,:-3]
y_dev1 = data_dev[:,36:]
X_test1 = data_dev[:,:-3]
y_test1 = data_dev[:,36:]
X_train1.shape,y_train1.shape
```

```python
class SoftMax_Reg():
    def __init__(self,epoch, lr):
        self.epoch = epoch
        self.lr = lr
        self.w = None
        self.train_loss = []
        self.val_loss = []
        self.metrics = []
    def stable_softmax(self,z):
        z_max = np.max(z, axis = 1, keepdims=True)
        exp_z = np.exp(z - z_max)
        return exp_z / np.sum(exp_z, axis= 1, keepdims=True)
    def predict(self,X):
        z = X @ self.w
        y_hat = self.stable_softmax(z)
        return y_hat
    def accuracy(self,y,y_predict):
        max_y = np.max(y_predict,axis=1,keepdims=True)
        y_classify = np.where(y_predict==max_y,1,0)
        return (y==y_classify).mean()
    def loss_func(self,y,y_hat):
        return -((y*np.log(y_hat + 1e-6)).sum(axis= 1)).mean()
    def fit(self,X_train,X_dev,y_train,y_dev):
        n = X_train.shape[0]
        d = X_train.shape[1]
        self.w = np.zeros((d,y_train.shape[1]))
        for i in range(self.epoch):
            y_hat = self.predict(X_train)
            train_loss = self.loss_func(y_train,y_hat)

            y_dev_hat = self.predict(X_dev)
            val_loss = self.loss_func(y_dev,y_dev_hat)

            acc = self.accuracy(y_train,y_hat)

            grad = (X_train.T @ (y_hat-y_train)) / n
            self.w -= self.lr * grad
            self.train_loss.append(train_loss)
            self.val_loss.append(val_loss)
            self.metrics.append(acc)
            print(f'\nEPOCH {i + 1}:\tTraining loss: {train_loss:.3f}\tValidation loss: {val_loss:.3f}')
        
```

```python
sm = SoftMax_Reg(1000,0.0001)
sm.fit(X_train1,X_dev1,y_train1,y_dev1)
```

```python
y_test_predict1 = sm.predict(X_test1)
test_accuracy1 = sm.accuracy(y_test1,y_test_predict1)
print(test_accuracy1)
```

```python
plt.figure(figsize=(10,5))
plt.plot(list(range(1, sm.epoch+1)), sm.train_loss, color = 'red' , label = 'loss')
plt.plot(list(range(1, sm.epoch+1)), sm.val_loss, color = 'blue' , label = 'val loss')
plt.xlabel('Epoch') 
plt.ylabel('Loss')
plt.title('Loss over epochs in training')
plt.legend()
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.plot(list(range(1, sm.epoch+1)), sm.metrics, color = 'red' , label = 'Accuracy')
plt.xlabel('Epoch') 
plt.ylabel('Accuracy')
plt.title('Accuracy over epochs in training')
plt.legend()
plt.grid()
plt.show()
```
